### what is the problem of the above codes?

```python
from bs4 import BeautifulSoup
# find title + text
title = []
para = {}
soup = BeautifulSoup(r.text, 'html.parser')
for elem in soup.find_all(['div', 'p']):
    if elem.name == 'div' and 'mw-heading' in elem.get('class', []):
      #print(elem.get_text())
      if elem.get_text() != '' and elem.get_text() not in title:
        #print(elem.get_text())
        cur_title = elem.get_text()
        title.append(cur_title )
        para.setdefault(cur_title, [])
    if elem.name == 'p':
      if len(title)  == 0:
        para.setdefault('no_title', []).append(elem.get_text())
      if len(title) != 0:
        para.setdefault(title[-1], []).append(elem.get_text())




```
